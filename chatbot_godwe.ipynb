```python
import google.generativeai as genai
import os

print("--Olá, Seja bem vindo ao seu chatbot para entender sobre energia! Faça uma pergunta ou encerre digitando 'sair'.")
genai.configure(api_key=os.environ['Gemini_API_IA'])

instrucao_inicial = """Você é um analista de produtos da goodWe, uma empresa de tecnologia que visa o desenvolvimento sustentavel e
esta com carregadores para carros eletricos. Este projeto tem que ter: Sistema que, com base na infraestrutura elétrica disponível do estabelecimento
(rede + possível geração local), realiza a distribuição inteligente da energia para os carregadores, evitando sobrecarga e reduzindo custos com upgrade elétrico.
O cálculo deve ser realizado de forma dinâmica, considerando:
Consumo total do estabelecimento em tempo real
Capacidade máxima contratada
Quantidade de veículos conectados
Prioridade de carregamento por perfil de usuário
Estratégia de distribuição de potência:
O primeiro veículo conectado recebe uma potência base inicial
Conforme novos veículos são conectados, o sistema redistribui a potência disponível entre todos os carregadores
O sistema deve reduzir ou aumentar a potência de forma automática com base:
No consumo geral do comércio
No horário (pico vs. fora de pico)
No nível de carga dos veículos
Otimizações recomendadas:
Priorização de usuários (ex: clientes premium, urgência de carga)
Redução gradual de potência para veículos próximos da carga completa
Corte ou limitação automática em horários de pico do estabelecimento
"""

model = genai.GenerativeModel(
    model_name="gemini-2.5-flash",
    system_instruction = instrucao_inicial
)

chat = model.start_chat(history=[])

while True:
  Pergunta = input("")
  if Pergunta.lower() == 'sair':
    break

  try:
    response = chat.send_message(Pergunta)

    model_response_content = response.text
    print(f"\n {model_response_content}")

  except Exception as e:
    print(f"\n Houve um erro ao obter a resposta do modelo: {e}")
```

```python
# Esta linha exibe uma mensagem de boas-vindas para o usuário quando o programa é iniciado.
# É útil para guiar o usuário sobre como interagir com o chatbot.
print("--Olá, Seja bem vindo ao seu chatbot para entender sobre energia! Faça uma pergunta ou encerre digitando 'sair'.")

# Configura a biblioteca 'genai' com a chave da API.
# A chave da API é recuperada da variável de ambiente 'Gemini_API_IA'.
# Isso é crucial para autenticar suas requisições à API do Gemini.
# Variável Escolhida: `Gemini_API_IA` - Nome da variável de ambiente onde a chave da API do Gemini está armazenada.
genai.configure(api_key=os.environ['Gemini_API_IA'])

# Esta string define o papel e as diretrizes do chatbot, instruindo-o a atuar como um analista de produtos da GoodWe.
# Contém detalhes sobre o projeto da GoodWe e os requisitos técnicos do sistema de carregamento.
# Esta é a 'persona' do seu assistente de IA.
# Variável Escolhida: `instrucao_inicial` - Um nome descritivo para indicar que é a instrução que define o comportamento inicial do sistema (chatbot).
instrucao_inicial = """Você é um analista de produtos da goodWe, uma empresa de tecnologia que visa o desenvolvimento sustentavel e
esta com carregadores para carros eletricos. Este projeto tem que ter: Sistema que, com base na infraestrutura elétrica disponível do estabelecimento
(rede + possível geração local), realiza a distribuição inteligente da energia para os carregadores, evitando sobrecarga e reduzindo custos com upgrade elétrico.
O cálculo deve ser realizado de forma dinâmica, considerando:
Consumo total do estabelecimento em tempo real
Capacidade máxima contratada
Quantidade de veículos conectados
Prioridade de carregamento por perfil de usuário
Estratégia de distribuição de potência:
O primeiro veículo conectado recebe uma potência base inicial
Conforme novos veículos são conectados, o sistema redistribui a potência disponível entre todos os carregadores
O sistema deve reduzir ou aumentar a potência de forma automática com base:
No consumo geral do comércio
No horário (pico vs. fora de pico)
No nível de carga dos veículos
Otimizações recomendadas:
Priorização de usuários (ex: clientes premium, urgência de carga)
Redução gradual de potência para veículos próximos da carga completa
Corte ou limitação automática em horários de pico do estabelecimento
"""

# Inicializa o modelo generativo com o nome 'gemini-2.5-flash'.
# O 'system_instruction' define o comportamento base do modelo para todas as interações.
# Variável Escolhida: `model` - Representa o modelo de IA que será utilizado para gerar as respostas.
model = genai.GenerativeModel(
    model_name="gemini-2.5-flash",
    system_instruction = instrucao_inicial
)

# Inicia uma nova sessão de chat com o modelo.
# A lista `history=[]` garante que a conversa comece do zero.
# O objeto `chat` gerencia automaticamente o histórico da conversa entre o usuário e o modelo.
# Variável Escolhida: `chat` - Representa a instância da sessão de chat, que mantém o estado da conversa e permite enviar mensagens e receber respostas.
chat = model.start_chat(history=[])

# Este é o loop infinito que mantém o chatbot ativo, aguardando a entrada do usuário.
while True:
  # Solicita a entrada do usuário. O texto vazio indica que o prompt será o cursor piscando.
  # Variável Escolhida: `Pergunta` - Armazena a string que o usuário digitou como sua pergunta ou comando.
  Pergunta = input("")

  # Verifica se o usuário digitou 'sair' (case-insensitive) para encerrar o chat.
  if Pergunta.lower() == 'sair':
    break

  # Bloco try-except para lidar com possíveis erros durante a comunicação com a API.
  try:
    # Envia a pergunta do usuário para o modelo através da sessão de chat.
    # O método `send_message` gerencia o histórico da conversa automaticamente.
    # Variável Escolhida: `response` - Contém o objeto de resposta retornado pela API do Gemini, que inclui o texto gerado e outras informações.
    response = chat.send_message(Pergunta)

    # Extrai o texto da resposta do modelo.
    # Variável Escolhida: `model_response_content` - Armazena apenas o conteúdo textual da resposta do modelo, que será exibido ao usuário.
    model_response_content = response.text

    # Imprime a resposta do modelo no console.
    print(f"\n {model_response_content}")

  # Captura qualquer exceção que ocorra durante o envio da mensagem ou recebimento da resposta.
  except Exception as e:
    # Imprime uma mensagem de erro amigável, incluindo a descrição do erro.
    print(f"\n Houve um erro ao obter a resposta do modelo: {e}")
```
