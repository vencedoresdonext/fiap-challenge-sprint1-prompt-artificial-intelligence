```python
import os
from openai import OpenAI
from google.colab import userdata

os.environ['OPENAI_API_KEY'] = userdata.get('OPENAI_API_KEY')

client = OpenAI()
```

```python
print("--Olá, Seja bem vindo ao seu chatbot de energia! Digite 'sair' para encerrar.\n")

msg = [{
    "role":"system"
    ,"content":'''Você é especialista em sistemas energéticos da GoodWe.

Objetivo:
Gerenciar energia para carregadores de veículos elétricos.

Regras:
- Não exceder capacidade elétrica
- Distribuir energia entre veículos
- Priorizar usuários quando necessário
- Considerar horário e energia solar

Responda de forma técnica, direta.
'''
}]

pergunta = ''

while True:
  pergunta = input()
  if pergunta.lower() == 'sair':
    break
  msg.append({
      "role": "user"
      , "content": pergunta
  })
  resposta = client.chat.completions.create(
      model="gpt-4o-mini"
      , messages=msg
  )

  print(f"\n{resposta.choices[0].message.content}") # Corrected from .messages to .message
  msg.append(
      {
          "role": "assistant"
          , "content": resposta.choices[0].message.content # Corrected from .messages to .message
      }
  )
```
